### Helper function to show image.

io is utility function to read and write different formats of images.

```python
def show(image):
    io.imshow(image)
    io.show()
```

Calling function show to display image.

```python
#get image of checkerboard in a variable
image_checkerboard = data.checkerboard()

#calling our helper function to display image by passing 
#image_checkerboard to show() function
show(image_checkerboard)
```

## Manipulating r,g,b of an image

Get image of an astronaut.

```python
#get image of an astronaut from data
image_ast = data.astronaut()

#Display shape of image_ast
image_ast.shape
```

```python
#Display shape of image_ast
show(image_ast)
```

Extracting r,g,b from an image.

```python
r = image_ast[:,:,0] #red   channel
g = image_ast[:,:,1] #green channel
b = image_ast[:,:,2] #blue  channel
```

Shape of all channel remains equal to the shape of image grid.

```python
(r.shape == g.shape == b.shape)
```

Lets play with r,g,b of astronaut image

```python
#setting r channel of image to 0
image_ast[:,:,0] = 0
show(image_ast)
```

```python
#setting g channel of image to 0
image_ast[:,:,1] = 0
show(image_ast)
```

```python
#setting b channel of image to 0
image_ast[:,:,2] = 0
show(image_ast)
```

Boolean masking getting r channel where r is greater than its mean. 

```python
r > r.mean()
```

```python
#Setting boolean mask to a variable
image_r = r > r.mean()

#Display image
show(image_r)
```

```python
image_g = g > g.mean()
show(image_g)
```

```python
image_b = b > b.mean()
show(image_b)
```

Coverting image to greyscale by using skimage.color.rgbtogrey.

```python
show(skimage.color.rgb2grey(image_ast))
```
